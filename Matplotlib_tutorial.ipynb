```python
days= [1,2,3,4,5,6,7]
max_t = [30,24,28,30,29,26,32]
avg_t = [32,23,24,32,24,27,30]
min_t = [33,25,27,36,24,23,32]
plt.plot(days, max_t)
plt.plot(days, avg_t)
plt.plot(days, min_t)
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.plot(days, max_t, label = "max")
plt.plot(days, avg_t, label = "average")
plt.plot(days, min_t, label = "min")
plt.legend()
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.plot(days, max_t, label = "max")
plt.plot(days, avg_t, label = "average")
plt.plot(days, min_t, label = "min")
plt.legend()
plt.grid()
```
